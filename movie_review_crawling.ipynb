```python
from bs4 import BeautifulSoup

html_doc = """
<html><head><title>The Dormouse's story</title></head>
<body>
<p class="title"><b>The Dormouse's story</b></p>
<p class="story">Once upon a time there were three little sisters;
and their names were
<a href="https://example.com/elsie" class="sister" id="link1">Elsie</a>
<a href="https://example.com/lacie" class="sister" id="link2">Lacie</a>
<a href="https://example.com/tillie" class="sister" id="link3">Tillie</a>;
and they lived at the bottom of a well.</p>

<p class="story">...</p>
"""

#Beautifulsoup 객체를 생성한다.
soup = BeautifulSoup(html_doc, 'lxml')


print("============객체.태그")
#객체.태그 이름
# .태그 이름으로 하위 태그로의 접근이 가능
print("soup.body.p의 결과 : ", soup.body.p)


print("==============객체.태그['속성이름']")
#객체.태그['속성 이름']
# 객체의 태그 속성은 파이썬 딕셔너리처럼 태그['속성 이름']으로 접근 가능
print("soup.a['href']의 결과: ", soup.a['href'])


print("================객체.name")
#객체.name
#name 변수
print("soup.title.name의 결과: ", soup.title.name)

print("============객체.contents")
#객체.contents
# 태그의 자식들을 리스트로 변환
print("soup.contents의 결과: ", soup.contents)


print("============find()")
# find() : 태그 하나만 가져옴
'''
find(name, attrs, recursive, string, **kwargs)

[옵션]
name - 태그 이름
attrs - 속성(딕셔너리로)
recursive - 모든 자식 or 자식
string - 태그 안에 텍스트
keyword - 속성(키워드로)

(주의) class는 파이썬 예약어이므로 class_를 사용!
'''
print("soup.find()의 결과: ", soup.find('a', attrs={'class':'sister'}))


print("===========find_all()")
#find_all() : 해당 태그가 여러 개 있을 경우 한꺼번에 모두 가져온다.
'''
find_all(name, attrs, recursive, string, limit, **kwargs)

[옵션]
Limit - 몇 개까지 찾을 것인가? find_all()로 검색했을 때, 수천, 수만 개가 된다면, 시간이 오래 걸릴 것이다.
이 때 몇 개 까지만 찾을 수 있도록 제한을 둘 수 있는 인자이다.
'''
print("soup.find_all()의 결과: ", soup.find_all('a', limit=2))
```

```python
import requests
from bs4 import BeautifulSoup

# '스파이더맨:노 웨이 홈'의 네이버 영화 리뷰
url = "https://movie.naver.com/movie/bi/mi/review.naver?code=208077"
# html 소스 가져오기
res = requests.get(url)
```

```python
res.text
```

```python
# html 파싱
soup = BeautifulSoup(res.text, 'lxml')

# 리뷰 리스트
ul = soup.find('ul', class_='rvw_list_area') # ul로 묶여있는 모든 리뷰의 제목
lis = ul.find_all('li') # li 로 묶인 각각의 리뷰 제목들을 리스트로 반환

len(lis)
```

```python
i = 0
for li in lis:
    i = i + 1
    print("[%d] %s"%(i, li.a.text))
```
